```python
import pandas as pd
import numpy as np
import folium
pd.options.mode.use_inf_as_na = True
#df.apply(lambda x: x) 
```

In the next cell, first, canada.xlsx file is loaded using pandas read_excel, then it is assigned to the dataframe "df_canada"
Then we set the not assigned values to numpy null values to later delete its rows
then the pandas function dropna is used to delet every row that contains np.nan values

```python
df_canada=pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/datascience-donotdelete-pr-moxy68pzdn0xi6/canada.xlsx?response-content-disposition=attachment%3B%20filename%3D%22canada.xlsx%22&response-content-type=application%2Fvnd.openxmlformats-officedocument.spreadsheetml.sheet&X-Amz-Algorithm=AWS4-HMAC-SHA256&X-Amz-Date=20200729T093747Z&X-Amz-SignedHeaders=host&X-Amz-Expires=86400&X-Amz-Credential=5e79306a71044e398886c24795880572%2F20200729%2Fus-geo%2Fs3%2Faws4_request&X-Amz-Signature=98424a2949361a27109f5dd71a286bc44165648fc5307f23880230ead4fde872')
#df_canada.reset_index(inplace = True)
df_canada['Borough']=df_canada['Borough'].replace('Not assigned', np.nan)
df_canada['Neighbourhood']=df_canada['Neighbourhood'].replace('Not assigned', np.nan)
df_canada.dropna(axis=0,inplace=True)
```

A DataFrame is created grouping by Postal code, in order to merge neighbourhoods with the same borough

```python
df_pc=df_canada.groupby(df_canada['Postal Code'])
df_pc=df_pc.apply(lambda x: x) 
df_pc['Borough']=df_pc['Borough'].replace('Not assigned', np.nan)
df_pc['Neighbourhood']=df_pc['Neighbourhood'].replace('Not assigned', np.nan)
df_pc.dropna(axis=0,inplace=True)
df_pc.set_index('Postal Code', inplace = True)
df_pc
```

The csv is read and assigned to the data frame named df_geo

```python
df_geo=pd.read_csv('https://cocl.us/Geospatial_data')

df_geo.set_index('Postal Code', inplace = True)
df_geo
```

The postal codes from the data frame, after data wrangling are assigned to the dataframe df_postalcode to extract latitude and longitude from df_geo by using merge function

```python
df_postalcode=pd.DataFrame(df_pc.iloc[:,0])
df_postalcode
df_canada_ln=pd.DataFrame(df_postalcode.merge(df_geo, left_on='Postal Code', right_on='Postal Code', how='inner'))
df_canada_ln
```

a new data frame is created named df_cn, that extracts some columns from the needed data

```python
df_cn=pd.DataFrame()
df_cn['Postal Code']=df_canada_ln.iloc[:,0]
df_cn['Borough']=df_pc['Borough']
df_cn['Neighbourhood']=df_pc['Neighbourhood']
df_cn['Latitude']=df_canada_ln['Latitude']
df_cn['Longitude']=df_canada_ln['Longitude']
df_cn.shape
```

```python
df_cn.shape
df_cn.drop(['Postal Code'],axis=1, inplace=True)
df_cn
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(df_cn['Borough'].unique()),
        df_cn.shape[0]
    )
)
print('The dataframe has {} postal codes and {} neighborhoods.'.format(
        len(df_cn.iloc[:,0].unique()),
        df_cn.shape[0]
    )
)
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_cn['Latitude'], df_cn['Longitude'], df_cn['Borough'], df_cn['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
